```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.cricbuzz.com/cricket-series/9325/icc-champions-trophy-2025/matches'
page=requests.get(url)
```

```python
Soup=BeautifulSoup(page.text,'html')
```

```python
content=Soup.find('div',class_='cb-bg-white cb-col-100 cb-col cb-hm-rght cb-series-filters').find_all('div',class_='cb-col-100 cb-col cb-series-brdr cb-series-matches')
```

```python
full_link_list=[]
match_summary=[]
for ele in content:
    Venues=ele.find('div',class_='text-gray').text
    Date=ele.find('div',class_='cb-col-25 cb-col pad10 schedule-date').find('span',class_='ng-binding')
    Matches=ele.find('div',class_='cb-col-60 cb-col cb-srs-mtchs-tm').find('span').text
    Result=ele.find('div',class_='cb-col-60 cb-col cb-srs-mtchs-tm').find('a',class_='cb-text-complete').text
    Time=ele.find('div',class_='cb-col-40 cb-col cb-srs-mtchs-tm').find('div',class_='cb-font-12 text-gray').text
    Links=ele.find('div',class_='cb-col-60 cb-col cb-srs-mtchs-tm').find('a',class_='text-hvr-underline').attrs['href']
    base_url = "https://www.cricbuzz.com"
    full_link = base_url + Links
    full_link_list.append(full_link)
    match_summary.append([Venues,Date,Matches,Result,Time,full_link])
    print(full_link)
    print(Time)
    print(Result)
    print(Matches)
    print(Date)
    print(Venues)
```

```python
full_scoreboard_list=[]
def commentary(link):
    url=link
    page=requests.get(url)
    Soup=BeautifulSoup(page.text,'html')
    content=Soup.find('nav',class_='cb-nav-bar')
    
    scorebord_link=content.find_all('a',class_='cb-nav-tab')
    
    if len(scorebord_link) > 1:  # Ensure there is a second <a> element
        second_a_tag = scorebord_link[1]  # Get the second <a> element
        if 'href' in second_a_tag.attrs:  # Ensure 'href' exists
            full_link = base_url + second_a_tag['href']
            full_scoreboard_list.append(full_link)
            print(full_link)  # Print only the second link
    else:
        print("Second link not found!")
    
    
```

```python
for link in full_link_list:
    commentary(link)
```

```python
scores_finals1=[]
first_innings_summary=[]
Player_History_Link=[]
def scoreboardfirstinn(link):
    i=0
    url=link
    page=requests.get(url)
    Soup=BeautifulSoup(page.text,'html')
    innings_1 = Soup.find('div', id='innings_1')  # First, check if it exists
    if not innings_1:
        print(f"Warning: 'innings_1' not found in {url}")
        return  # Exit function if not found

    content = innings_1.find('div', class_='cb-col cb-col-100 cb-ltst-wgt-hdr')
    if not content:
        print(f"Warning: No matching div found for {url}")
        return  # Exit function if content is missing
    
    score_country=innings_1.find('div',class_='cb-col cb-col-100 cb-scrd-hdr-rw').find('span').text
    print(score_country)
    for ele in content.find_all('div', class_='cb-col cb-col-100 cb-scrd-itms'):
        # Find all matching divs
        batsman = ele.find('div', class_="cb-col cb-col-25")  # Find the inner div
        if batsman:  # Ensure it is not None
            batsman_name = batsman.text.strip()
            i=i+1# Extract text
            batsman_profile=batsman.find('a').attrs['href']
            base_url = "https://www.cricbuzz.com"
            full_link_pr = base_url +  batsman_profile
            print(full_link_pr)
        outornot=ele.find('div',class_='cb-col cb-col-33')
        if outornot:  # Ensure it is not None
            outornot_detail = outornot.text.strip() 
            # Extract text
        stats = ele.find_all('div', class_="cb-col cb-col-8 text-right text-bold") + \
        ele.find_all('div', class_="cb-col cb-col-8 text-right") + \
        ele.find_all('div', class_="cb-col cb-col-8 text-right", style=True)  # Include styled elements

        if stats:
            extracted_stats = [stat.text.strip() for stat in stats]

    # Ensure we have at least 5 elements
            while len(extracted_stats) < 5:
                extracted_stats.append("N/A")

            runs, Bowls, Fours, Sixes, Strike_rate = extracted_stats[:5]
        else:
            extracted_stats = []

        print(runs, Bowls, Fours, Sixes, Strike_rate)
        Player_History_Link.append(full_link_pr)
        #second_innings_summary.append([bowler_name,overs, maidens,runs_given,wickets,economy])
        
        first_innings_summary.append([i,batsman_name,score_country,outornot_detail,runs, Bowls, Fours, Sixes, Strike_rate,url])
    scores_finals1.append(score_country)
```

```python
Player_summary=[]
def playerdetails(link):
    url=link
    page=requests.get(url)
    Soup=BeautifulSoup(page.text,'html')
    content = Soup.find('div', class_='cb-col cb-col-100 cb-bg-white')  # First, check if it exists
    if not content:
        print(f"Warning: 'content' not found in {url}")
        return  # Exit function if not found
    
    if content:
        image_link=content.find('img',title='profile image').attrs['src']
        if image_link:
            print(image_link)
        else:
            print("Image not found")
        name=content.find('div',class_='cb-col cb-col-80 cb-player-name-wrap').find('h1').text
        country=content.find('div',class_='cb-col cb-col-80 cb-player-name-wrap').find('h3').text
        print(name,country)
    
    hist_det=Soup.find('div',class_='cb-col cb-col-33 text-black')
    if hist_det:
        history=hist_det.find_all('div',class_='cb-col cb-col-60 cb-lst-itm-sm')
        
        birth=history[0].text.strip()
        birth_place=history[1].text.strip()
        role=history[2].text.strip()
        batting_style=history[3].text.strip()
        bowling_style=history[4].text.strip()
        print(birth, birth_place,role,batting_style,bowling_style)
    Player_summary.append([name,country,image_link,birth, birth_place,role,batting_style,bowling_style])
```

```python
for link in unique_Player_History_Link:
    playerdetails(link)
```

```python
player_headers=['name','country','Player_summary','birth', 'birth_place','role','batting_style','bowling_style']
df=pd.DataFrame(Player_summary,columns=player_headers)
df
df.to_csv(r'D:\NASMEEN\University\Data Science\CT 2025\New CSV\Player_Summarry.csv',index=False)
```
